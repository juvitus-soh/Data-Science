```python

plt.scatter(data['longitude'], data['latitude'], alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot of Berlin Properties')
plt.show()
```

The scatter plot shows a high concentration of homes between latitudes 52.45 and 52.55, and longitudes 13.3 and 13.5. The distribution of homes becomes saprtial as we move further awaye

```python
plt.hist(data["room_type"])
plt.xlabel("Room type")
plt.ylabel("Number of properties")
plt.show()
```

The most common type available are Private rooms, followed by Entire home/spt and then next on a very low level is shared room

OTHER ANALYSIS

```python
df = data.dropna(subset=['price'])
# Histogram of property prices
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=50, edgecolor='black')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.title('Histogram of Property Prices')
plt.show()

# Box plot of property prices
plt.figure(figsize=(10, 6))
plt.boxplot(df['price'], vert=False)
plt.xlabel('Price ($)')
plt.title('Box Plot of Property Prices')
plt.show()
```
